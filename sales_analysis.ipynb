```python
import pandas as pd 
// Load the dataset
df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
# Display the first few rows of the dataframe.
df.head()
```

```python
#Display the summary information of the dataframe.
df.info()
```

```python
# Display statistical summary of the dataframe.
df.describe()
```

```python
# Display statistical summary of all columns including categorical ones.
df.describe(include='all')
```

```python
# Check for missing values in the dataframe.
df.isnull().sum()
```

```python
# Fill missing values in 'Review Rating' column with the median rating of the respective 'Category'.
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
# Check for missing values in the dataframe.
df.isnull().sum()
```

```python
# Standardize column names: lowercase.
df.columns = df.columns.str.lower()
# Replace spaces with underscores in column names.
df.columns = df.columns.str.replace(' ', '_')
```

```python
# Display the column names of the dataframe.
df.columns
```

```python
df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'}, inplace=True)
df.columns
```

```python
# create a column age_group based on age
labels = ['Young Adult', 'Adult', 'Middle Aged', 'Senior']
# Create age_group using quantiles
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
# Display the first 10 rows of age and age_group columns.
df[['age','age_group']].head(10)
```

```python
df[['previous_purchases', 'frequency_of_purchases']].head(10)
```

```python
# Map frequency_of_purchases to numerical values representing days.
frequency_mapping={
    
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Annually':365,
    'Fortnightly':14,
    'Bi-Weekly':14,
    'Every 3 Months':90,
}
# Create a new column 'purchase_frequency_days' based on the mapping.
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping) 
```

```python
# Fill NaN values with -1 before converting to int.
df['purchase_frequency_days'] = df['purchase_frequency_days'].fillna(-1).astype(int)
df[['frequency_of_purchases', 'purchase_frequency_days']].head(10)
```

```python
df[['discount_applied', 'promo_code_used']].head(10)
```

```python
# Check if 'discount_applied' matches 'promo_code_used'
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# If they match, we can drop one of the columns to avoid redundancy.
df.drop(columns=['promo_code_used'], inplace=True)
```
